```python
import json
import pandas as pd

# Load F1 scores
with open('results/f1_scores_per_threshold.json', 'r') as f:
    f1_scores = json.load(f)

f1_scores_df = pd.DataFrame(f1_scores)#.astype(float)
# Create separate DataFrames for f1, precision, and recall
f1_df = pd.DataFrame({metric: {float(threshold): data['f1'] for threshold, data in metric_data.items()} 
                      for metric, metric_data in f1_scores.items()})

precision_df = pd.DataFrame({metric: {float(threshold): data['precision'] for threshold, data in metric_data.items()} 
                             for metric, metric_data in f1_scores.items()})

recall_df = pd.DataFrame({metric: {float(threshold): data['recall'] for threshold, data in metric_data.items()} 
                          for metric, metric_data in f1_scores.items()})

# Sort the index (thresholds) of each DataFrame
f1_df.sort_index(inplace=True)
precision_df.sort_index(inplace=True)
recall_df.sort_index(inplace=True)

f1_scores_df
```

```python
f1_df.max().nlargest(10)
```

```python
# Find the maximum entry in f1_scores_df
# Find the top 10 F1 scores
top_10_f1_scores = f1_df.max().nlargest(10)

print("Top 10 F1 Scores:")
for metric, score in top_10_f1_scores.items():
    threshold = f1_df[metric].idxmax()
    print(f"Metric: {metric}")
    print(f"F1 Score: {score:.6f}")
    print(f"Threshold: {threshold}")
    print()

# Still keep the overall maximum for reference
max_f1_score = top_10_f1_scores.max()
max_f1_metric = top_10_f1_scores.idxmax()
max_f1_threshold = f1_df[max_f1_metric].idxmax()

print(f"Overall Maximum F1 Score: {max_f1_score:.6f}")
print(f"Metric: {max_f1_metric}")
print(f"Threshold: {max_f1_threshold}")
```

```python
# Find entries in precision_df > 0.9, then find corresponding max recall
high_precision_thresholds = {}

for metric in precision_df.columns:
    # Find thresholds where precision > 0.9
    high_precision_mask = precision_df[metric] > 0.9
    
    if high_precision_mask.any():
        # Get corresponding recall values
        corresponding_recalls = recall_df.loc[high_precision_mask, metric]
        
        # Find the threshold with max recall
        max_recall_threshold = corresponding_recalls.idxmax()
        
        high_precision_thresholds[metric] = {
            'threshold': max_recall_threshold,
            'precision': precision_df.loc[max_recall_threshold, metric],
            'recall': recall_df.loc[max_recall_threshold, metric],
            'f1': f1_df.loc[max_recall_threshold, metric]
        }

# Display results
print("Metrics with high precision (>0.9) and maximum corresponding recall:")
for metric, values in high_precision_thresholds.items():
    print(f"\n{metric}:")
    print(f"  Threshold: {values['threshold']:.2f}")
    print(f"  Precision: {values['precision']:.4f}")
    print(f"  Recall: {values['recall']:.4f}")
    print(f"  F1 Score: {values['f1']:.4f}")

# Find the metric with the highest recall among high precision results
best_metric = max(high_precision_thresholds, key=lambda x: high_precision_thresholds[x]['recall'])
best_values = high_precision_thresholds[best_metric]

print(f"\nBest metric maintaining precision > 0.9 with highest recall:")
print(f"Metric: {best_metric}")
print(f"Threshold: {best_values['threshold']:.2f}")
print(f"Precision: {best_values['precision']:.4f}")
print(f"Recall: {best_values['recall']:.4f}")
print(f"F1 Score: {best_values['f1']:.4f}")
```
